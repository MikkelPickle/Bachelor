```python
f2 = open("vector_file2.txt", "r")
differences2 = [int(line.strip('\n')) for line in f2]
```

```python
l1normm = 14798569
l1norm = 13927067
l2norm = 954818
N = len(differences)
eps = 0.01
eps_prime = np.sqrt(eps*2)
delta = 0.10
factor = 0.2
percentile = [i/N for i in range(1,N+1)]
```

```python
#Count Min Sketch
plt.figure(figsize=(12, 8))
plt.plot(percentile, differences, label='Count Min Sketch', color = "purple", linestyle = '--', linewidth = 2)
plt.plot(percentile, differences2, label='Count Sketch', linestyle = '--', linewidth = 2)
plt.xlim(0, 1.05)
plt.ylim(0, 200000)
plt.axhline(y=eps*l1norm, color='red', linestyle='-', label = r"$\epsilon \cdot \ell 1$-norm")
plt.axhline(y=eps_prime*l2norm, color='orange', linestyle='-', label = r"$\epsilon' \cdot \ell 2$-norm")
plt.axvline(x=1-delta, color='green', linestyle='-', label = r"$1-\delta$")
plt.title("Performance of Algorithms")
plt.xlabel("Percentile of Errors")
plt.ylabel("Approximation Error")
plt.legend()
plt.show()
```
